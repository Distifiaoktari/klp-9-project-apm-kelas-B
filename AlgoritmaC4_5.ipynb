```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("LoanApplicantData.csv", sep=',')
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
dataset['Loan_ID'] = enc.fit_transform(dataset['Loan_ID'].values)
dataset['Gender'] = enc.fit_transform(dataset['Gender'].values)
dataset['Married'] = enc.fit_transform(dataset['Married'].values)
dataset['Dependents'] = enc.fit_transform(dataset['Dependents'].values)
dataset['Education'] = enc.fit_transform(dataset['Education'].values)
dataset['Self_Employed'] = enc.fit_transform(dataset['Self_Employed'].values)
dataset['ApplicantIncome'] = enc.fit_transform(dataset['ApplicantIncome'].values)
dataset['CoapplicantIncome'] = enc.fit_transform(dataset['CoapplicantIncome'].values)
dataset['LoanAmount'] = enc.fit_transform(dataset['LoanAmount'].values)
dataset['Loan_Amount_Term'] = enc.fit_transform(dataset['Loan_Amount_Term'].values)
dataset['Credit_History'] = enc.fit_transform(dataset['Credit_History'].values)
dataset['Property_Area'] = enc.fit_transform(dataset['Property_Area'].values)
dataset['Loan_Status'] = enc.fit_transform(dataset['Loan_Status'].values)
```

```python
dataset
```

```python
atr_dataset = dataset.drop(columns = 'Loan_Status')
atr_dataset
```

```python
cls_dataset = dataset['Loan_Status']
cls_dataset
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(atr_dataset, cls_dataset, test_size=0.2, random_state=1)
tree_dataset =  DecisionTreeClassifier(random_state=1)
tree_dataset.fit(xtrain, ytrain)
```

```python
y_pred = tree_dataset.predict(xtest)
cm = confusion_matrix(ytest, y_pred)
print("Confusion Matrix")
print(cm)
akurasi = classification_report(ytest, y_pred)
print("Tingkat Akurasi Algoritma C4.5")
print("Akurasi :", akurasi)
akurasi = accuracy_score(ytest, y_pred)
print("Tingkat Akurasi: %d persen" %(akurasi*100))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree_dataset, out_file="Tree_Loanstatus.dot", class_names=["Y","N"],
                feature_names=atr_dataset.columns, impurity=False, filled=True)
```
